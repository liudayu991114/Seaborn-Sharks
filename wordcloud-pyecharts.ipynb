```python
import pyecharts
import pandas as pd
```

```python
csvdata=pd.read_excel("dataset_final.xlsx")
csvdata
```

```python
import numpy as np
csvdata['sum_mental_level']=csvdata['Anxiety']+csvdata['Depression']+csvdata['Insomnia']+csvdata['OCD']
#high level 25-40 mid-level 13-26 low level 0-12
csvdata['sum_group']=csvdata['sum_mental_level'].apply(lambda x: 'low level' if x<=12 else 'mid level' if x<25 else 'high level')
csvdata[csvdata['sum_group']=='high level']
```

```python
high_level_group={}
mid_level_group={}
low_level_group={}
for clas in csvdata[csvdata['sum_group']=='high level'].groupby('Fav genre').groups.keys():
    high_level_group[clas]=len(csvdata[csvdata['sum_group']=='high level'].groupby('Fav genre').groups[clas])
for clas in csvdata[csvdata['sum_group']=='mid level'].groupby('Fav genre').groups.keys():
    mid_level_group[clas]=len(csvdata[csvdata['sum_group']=='mid level'].groupby('Fav genre').groups[clas])
for clas in csvdata[csvdata['sum_group']=='low level'].groupby('Fav genre').groups.keys():
    low_level_group[clas]=len(csvdata[csvdata['sum_group']=='low level'].groupby('Fav genre').groups[clas])
high_level_group_wordtuplelist=[(key,int(round(high_level_group[key]/len(csvdata[csvdata['sum_group']=='high level']),2)*100)) for key in high_level_group.keys()]
mid_level_group_wordtuplelist=[(key,int(round(mid_level_group[key]/len(csvdata[csvdata['sum_group']=='mid level']),2)*100)) for key in mid_level_group.keys()]
low_level_group_wordtuplelist=[(key,int(round(low_level_group[key]/len(csvdata[csvdata['sum_group']=='low level']),2)*100)) for key in low_level_group.keys()]
high_level_group_wordtuplelist
```

```python
from pyecharts import options as opts
from pyecharts.charts import WordCloud, Tab
from pyecharts.globals import SymbolType

def generate_wordcloud(data, color):
    c = (
        WordCloud()
        .add("", data, word_size_range=[30,150])#, shape=SymbolType.DIAMOND)
        .set_global_opts(title_opts=opts.TitleOpts(title="WordCloud for different mental health groups (%)"),
                        visualmap_opts=opts.VisualMapOpts(max_=33,min_=1,is_show=True))
        #.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}", color=color))
    )
    return c

def generate_interactive_wordcloud():
    tab = Tab()
    tab.add(generate_wordcloud(low_level_group_wordtuplelist, lambda x: "green"), "Low")
    tab.add(generate_wordcloud(mid_level_group_wordtuplelist, lambda x: "yellow"), "Mid")
    tab.add(generate_wordcloud(high_level_group_wordtuplelist, lambda x: "red"), "High")
    return tab

def save_interactive_wordcloud(filename):
    c = generate_interactive_wordcloud()
    c.render(filename)
    
#save html
save_interactive_wordcloud("interactive_wordcloud_2.html")
```
